```python
def plot_image_with_rectangles_and_text(img, 
                                        recs: list, 
                                        rec_color=(0, 255, 0), 
                                        text=None, 
                                        text_color=(0, 0, 255)):
    ploted_img = img.copy()
    for rec in recs:
        ploted_img = cv2.rectangle(ploted_img, (rec[0], rec[1]), (rec[2], rec[3]), rec_color, 1)
    if text:
        text_l = text.split('\n')
        for i in range(len(text_l)):
            t = text_l[i]
            ploted_img = cv2.putText(ploted_img, t, (5, 20 * (i + 1)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1)
    return ploted_img
```
